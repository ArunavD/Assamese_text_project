```python
from bs4 import BeautifulSoup
import requests
import urllib3
```

```python
url = input("Enter a website to extract the URL's from: ")
```

```python
ux = urllib3.PoolManager()
res = ux.request('GET', url)
#print(res)
soup = BeautifulSoup(res.data)
v = soup.body
print(v.ul.string)
print(v.li.string)
print(v.h2.string)
print(v.h3.string)
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```
