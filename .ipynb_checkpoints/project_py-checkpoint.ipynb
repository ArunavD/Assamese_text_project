```python
from bs4 import BeautifulSoup
import requests
import urllib3
import re
```

```python
url = input("Enter a website to extract the URL's from: ")
```

```python
ux = urllib3.PoolManager()
r = ux.request('GET', url)
#r  = ux.requests.get(url)
#data= r.text
soup = BeautifulSoup(r.data,'html.parser')
result=soup.body
for x in result.find_all('h3'):
    print(x.text)
```
