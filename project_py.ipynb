```python
from bs4 import BeautifulSoup
import requests
import urllib3
import re
```

```python
url = input("Enter a website to extract the URL's from: ")
```

```python
ux = urllib3.PoolManager()
res = ux.request('GET', url)
#print(res)
soup = BeautifulSoup(res.data,'html.parser')
v = soup.body
low = []
for x in soup.find_all('ul'):
    low.append(x.text)
for x in soup.find_all('li'):
    low.append(x.text)
for x in soup.find_all('h3'):#, 'h2', 'h3'):
    low.append(x.text)
print(low)

```

```python
#list of words
spl = low[0].split()
for i in range(1,len(low)):
    spl.append(spl.extend(low[i].split()))
print(spl)
```

```python
#removing english words from the list

def isEnglish(word):
    result = False
    english_min = u'\u0021'
    english_max = u'\u007F'
    result = False
    if not word:
        return False
    for character in word:
        if english_min <= character <= english_max:
            result = True
            break
    return result

#print([a for a in spl if not isEnglish(a)])

X = [a for a in spl if not isEnglish(a)]
print(X)
```

```python
#since [None, 'тАФ', 'тАФтАФ'] are still in the list, so to remove them_______

stopwords = [None,'тАФтАФ','тАУ']


for word in list(X):  
    if word in stopwords:
        X.remove(word)
print(X)
```

```python
#converting list to string

strX= "\n".join(X)
print(strX)
```

```python
# savings all data into a text file....

appendFile= open('dataset.txt','a')
appendFile.write('\n')
appendFile.write(strX)
appendFile.close()
```
