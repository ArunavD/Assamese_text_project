```python
from bs4 import BeautifulSoup
import requests
```

```python
url = input("Enter a website to extract the URL's from: ")
```

```python
r  = requests.get(url)
data= r.text
soup = BeautifulSoup(data)
print(soup.body)
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```
