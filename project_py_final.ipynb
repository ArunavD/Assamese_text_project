```python
from bs4 import BeautifulSoup
import requests
import urllib3
```

```python
url = input("Enter a website to extract the URL's from: ")
```

```python
ux = urllib3.PoolManager()
res = ux.request('GET', url)
#print(res)
soup = BeautifulSoup(res.data,'html.parser')
v = soup.body
for x in soup.find_all('ul'):
    print(x.text)
for x in soup.find_all('li'):
    print(x.text)
for x in soup.find_all('h3'):#, 'h2', 'h3'):
    print(x.text) 
```
